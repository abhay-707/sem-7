```python
# Function and its derivative
def f(x):
    return (x + 3)**2

def df(x):
    return 2 * (x + 3)

# Initial guess
x = 2

# Learning rate
alpha = 0.1

# Number of iterations
epochs = 50

print("Iteration\tX Value\t\tFunction Value")
for i in range(epochs):
    grad = df(x)
    x = x - alpha * grad
    print(f"{i+1}\t\t{x:.6f}\t\t{f(x):.6f}")

print("\nLocal minimum occurs at x =", round(x, 4))
```
